```python
import math
import cv2
import os
```

```python
cap = cv2.VideoCapture('data baru/ngantuk.mp4')
fps=math.floor(cap.get(cv2.CAP_PROP_FPS))
frame_count = 0
count=0
# Loop through each frame
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
      break
    else:
        if(frame_count % fps == 0):
            print()
            filename = os.path.join('data baru/ngantuk', f'ngantuk--{count}.jpg')
            cv2.imwrite(filename, frame)
            print(filename)
            count+=1
        frame_count+=1
cap.release()
cv2.destroyAllWindows()
```
