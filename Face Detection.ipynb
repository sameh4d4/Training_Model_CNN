```python
import cv2
import matplotlib.pyplot as plt
import os
import math
import numpy as np
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
def detect_faces(image):
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  # Detect faces in the image
  faces = face_cascade.detectMultiScale(gray, 1.3, 10)

  img1 = 0
  largest_face = None
  largest_area = 0
  for (x, y, w, h) in faces:
      area = w * h
      if area > largest_area:
          largest_area = area
          largest_face = (x, y, w, h)

  if largest_face is not None:
      x, y, w, h = largest_face
      cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
      face_roi = image[y:y + h, x:x + w]
      img1 = face_roi

  return img1
```

```python
def detect_faces2(image):
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  # Detect faces in the image
  faces = face_cascade.detectMultiScale(gray, 1.3, 1)

  img1 = 0
  largest_face = None
  largest_area = 0
  for (x, y, w, h) in faces:
      area = w * h
      if area > largest_area:
          largest_area = area
          largest_face = (x, y, w, h)

  if largest_face is not None:
      x, y, w, h = largest_face
      cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
      face_roi = image[y:y + h, x:x + w]
      img1 = face_roi

  return img1
```

```python
name_list_sadar=[]
name_list_ngantuk=[]
image_list_ngantuk=[]
image_list_sadar=[]
```

```python
for filename in os.listdir('data baru/ngantuk'):
  if filename.endswith(".jpg") or filename.endswith(".png"):
    image_path = os.path.join('data baru/ngantuk', filename)
    image = cv2.imread(image_path)
    image_list_ngantuk.append(image)
    name_list_ngantuk.append(filename)
```

```python
for filename in os.listdir('data baru/sadar'):
  if filename.endswith(".jpg") or filename.endswith(".png"):
    image_path = os.path.join('data baru/sadar', filename)
    image = cv2.imread(image_path)
    image_list_sadar.append(image)
    name_list_sadar.append(filename)
```

```python
image_list1 = image_list_ngantuk.copy()
no_face_detected_ids = []
for i, image in enumerate(image_list1):
  img_copy = image.copy()
  simpan=1
  face_img = detect_faces(img_copy)
  if face_img is 0:
    face_img = detect_faces2(img_copy)
    if face_img is 0:
      simpan=0
      no_face_detected_ids.append(i)
  if simpan == 1:
    print('saving ngantuk - '+name_list_ngantuk[i])
    cv2.imwrite(os.path.join('data baru/ngantuk', 'wajah/'+name_list_ngantuk[i]), face_img)
```

```python
image_list2 = image_list_sadar.copy()
no_face_detected_ids = []
for i, image in enumerate(image_list2):
  img_copy = image.copy()
  simpan=1
  face_img = detect_faces(img_copy)
  if face_img is 0:
    face_img = detect_faces2(img_copy)
    if face_img is 0:
      simpan=0
      no_face_detected_ids.append(i)
  if simpan == 1:
    print('saving sadar - '+name_list_sadar[i])
    cv2.imwrite(os.path.join('data baru/sadar', 'wajah/'+name_list_sadar[i]), face_img)
```

```python
def detect_biggest_face(image):
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  faces = face_cascade.detectMultiScale(gray, 1.3, 5)

  biggest_face = None
  biggest_area = 0

  for (x, y, w, h) in faces:
    area = w * h
    if area > biggest_area:
      biggest_area = area
      biggest_face = (x, y, w, h)

  if biggest_face is not None:
    (x, y, w, h) = biggest_face
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)
    face_roi = image[y:y+h, x:x+w]
    return face_roi
  else:
    return None
```

```python
for i, image in enumerate(image_list_ngantuk):
  face_roi = detect_biggest_face(image.copy())
  if face_roi is None:
    print(f"No face detected in image {i}")
  else:
    print('saving ngantuk - '+name_list_ngantuk[i])
    cv2.imwrite(os.path.join('data baru/ngantuk', 'wajah/'+name_list_ngantuk[i]), face_roi)
```

```python
for i, image in enumerate(image_list_sadar):
  face_roi = detect_biggest_face(image.copy())
  if face_roi is None:
    print(f"No face detected in image {i}")
  else:
    print('saving sadar - '+name_list_sadar[i])
    cv2.imwrite(os.path.join('data baru/sadar', 'wajah/'+name_list_sadar[i]), face_roi)
```
